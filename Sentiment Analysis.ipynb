## baru 1 oktober

```python
# Replace with your actual Bearer Token
bearer_token = "AAAAAAAAAAAAAAAAAAAAACU34gEAAAAAGPodDePavKJhtXIYWv48jDQUlI4%3DPx2PYLaYgSyfI1g8oVmIeHtXsJuLKNVFkhcKzK42XaeYNrhYUJ"
# Authenticate with Twitter API v2 using Bearer Token
client = tweepy.Client(bearer_token)

# Query dengan fokus pada opini & sentimen publik
query = '(Prabowo) (support OR vote OR pilih OR dukung OR like OR respect OR kecewa OR kritik OR corrupt OR hate OR gagal OR disappointed OR bangga OR pemimpin OR presiden OR MBG OR israel OR PBB OR gaza OR "makan bergizi gratis") lang:id -is:retweet'

# Ambil tweets (misalnya 50 dulu biar aman)
response = client.search_recent_tweets(
    query=query,
    max_results=80,
    tweet_fields=["id","text","created_at","lang","public_metrics","author_id"]
)

# Convert ke DataFrame
tweets_list = []
if response.data:
    for tweet in response.data:
        tweets_list.append([
            tweet.id,
            tweet.text,
            tweet.created_at,
            tweet.lang,
            tweet.author_id,
            tweet.public_metrics.get("like_count", 0),
            tweet.public_metrics.get("retweet_count", 0),
            tweet.public_metrics.get("reply_count", 0),
            tweet.public_metrics.get("quote_count", 0)
        ])

tweets_df = pd.DataFrame(tweets_list, columns=[
    "Tweet ID", "Tweet Text", "Created At", "Language", "Author ID",
    "Likes", "Retweets", "Replies", "Quotes"
])

# Tampilkan contoh data
display(tweets_df.head())
```

# Concat

```python

# Load ketiga dataset
df_new = pd.read_csv("prabowo_tweets.csv")
df_old = pd.read_csv("older_prabowo_tweets.csv")
df_older = pd.read_csv("lebih_older_prabowo_tweets.csv")
df_mbg = pd.read_csv("mbg_prabowo_tweets.csv")
df_mbg_lagi = pd.read_csv("mbg_lagi_prabowo_tweets.csv")
df_1oktober = pd.read_csv("tweet1oktober.csv")
df_1oktober_lagi = pd.read_csv("tweet1oktober_lagi.csv")
df_1oktober_coba_lagi = pd.read_csv("tweet1oktober_coba_lagi.csv")
df_mbg_aja = pd.read_csv("tweet1oktober_mbg.csv")

# Gabungkan semua
dataset = pd.concat([df_new, df_old, df_older, df_mbg, df_mbg_lagi,df_1oktober,df_1oktober_lagi,df_1oktober_coba_lagi, df_mbg_aja], ignore_index=True)

# Hapus duplikat (berdasarkan Tweet ID)
dataset.drop_duplicates(subset=["Tweet ID"], inplace=True)

# Reset index biar rapi
dataset.reset_index(drop=True, inplace=True)

# Simpan ke CSV
dataset.to_csv("prabowo_dataset.csv", index=False, encoding="utf-8")
```
